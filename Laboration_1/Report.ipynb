# Movie recommender

## Introduction

The goal for this project was to create and movie recommender that reccomends movies based on a input of a movie title. The objective of this assignment was to identify similarities between movies in order to be able to give relevant recommendations. Recommendation systems can often be divided into content-based methods and collabrative filtering. In this project both methods were used. Content-based filtering represented each movie as a feature vector coming from metadata. In this project TF-IDF was used. The similarity between the different movies is calculated using cosine similarity:
cos(0) = (A * B) / (||A|| ||B||)
Collabrative filtering was on the other hand uses user ratings and tries to learn different user behaviors.

## Data Analysis

This dataset consists of three different components:
-Movies(Titles and genres)
-Tags(User metadata)
-Ratings(User ratings)

Genres provide general categories for the movies while tags give a more in depth description of the movie. Therefore theese two were combined into a unified textual representation. The ratings dataset is very sparse. This means that many movies do not share overlapping users.  

## Model

### Several approaches were considered:
- Content-based filtering
- Collabrative filtering
- Hybrid model

The final model is a hybrid approach.
Genres and tags were combined into a single text feature per movie. This text was transformed into numerical vectors using TF-IDF. Cosine-similarity was then used to compute similarity between movies.
A small part of the ratings dataset was used to create a movie-user matrix. Missing values were filled with a zero. Yet again cosine-similarity was used to compute similratiy between movies based on user behavior.

The final recommendations score is a weighted combination of the two.
final_score = 0.6 * content_score + 0.4 * rating_score

### Hyperparameters:
- TF-IDF with english stopwords
- Ratings sample size set to 200_000
- Weights: Content = 0.6 and rating = 0.4



## Results

The model was evaluated using the input: Toy Story (1995) and the system produced the following recommendations:
- Toy Story 2 (1999)
- A Bug's Life (1993)
- Toy Story 3 (2010)
- Monsters , Inc. (2001)
- Finding Dory (2016)

The recomendations are very consistent and relevant as they share both similar themes and genres.
The rating component contributes to the final score when there are enough overlapping users but content-based component is clearly dominant.

## Discussion

One big limitation of the model is that collabrative filtering is based on a subset of the data. This does reduce the compution needed but really weakens the rating-based similarity.
The dataset is also sparse, which reduces the number of user overlaps between movies.
Despite theese limitiations the model preforms well and using the Hybrid approach is very effective due to the fact that it combines the strengths of both models.

## Conclusion

A hybrid recommendation system was sucsessfully implemented. The system generates five similar movies based metadata and on user behavior. The results demonstrate that the content-based method works very well while collabrative filtering can imporve performance with enough data.
